```python
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_val)
```

```python
def normalize_shap(I_c):

        
        sum_I_c = np.sum(I_c)
        I_c_normalized = I_c / sum_I_c
        return I_c_normalized

def clusterEntropy(model, X, y_pred):

    explainer = shap.TreeExplainer(model)
    shap_values = explainer.shap_values(X)
    epsilon = 1e-20
    N = len(y_pred)
    F = X.shape[1]
    total_entropy = 0

    for class_idx in np.unique(y_pred):
        class_shap_values = shap_values[class_idx][y_pred == class_idx]
        I_c = np.mean(np.abs(class_shap_values), axis=0)

        if np.all(I_c == 0):
            return 0
        else:
            
            I_c_prime = normalize_shap(I_c)
            class_entropy = -np.sum(I_c_prime * np.log2(I_c_prime + epsilon))

        total_entropy += (np.sum(y_pred == class_idx) * class_entropy)

    E_prime = total_entropy / (N * np.log2(F))
    return 1 - E_prime


val_simplicity = clusterEntropy(model, X_val, y_val)


```

```python

len_x = len(X_val.columns)
# Calculate significant SHAP counts for each class in a more concise way
significant_shap_counts = [np.sum(np.abs(values).mean(axis=0) / np.sum(np.abs(values).mean(axis=0)) > 0.01) for values in shap_values]

# Find the maximum count and calculate the percentage of significant SHAP values
total_significant_shap = max(significant_shap_counts)
percent_significant_shap = (total_significant_shap / len_x) * 100

print(f"Total significant shap: {total_significant_shap}")
print(f"Percent significant shap: {round(percent_significant_shap, 2)}%")
```

```python
import pandas as pd
import os

# Chemin du fichier CSV
filename = 'user_study_results.csv'

# Nouvelles données à ajouter
new_data = {
    "simplicity": [val_simplicity],
    "accuracy": [accuracy],
    "total_significant_shap": [total_significant_shap]
}

# Vérifie si le fichier existe pour lire l'identifiant le plus élevé
if os.path.exists(filename):
    # Lit le fichier existant
    df_existing = pd.read_csv(filename)
    
    # Obtient le dernier identifiant si le fichier n'est pas vide, sinon commence à 0
    last_id = df_existing['id'].max() if not df_existing.empty else 0
else:
    # Initialise un DataFrame vide avec les colonnes souhaitées si le fichier n'existe pas
    df_existing = pd.DataFrame(columns=['id'] + list(new_data.keys()))
    last_id = 0

# Crée un DataFrame avec la nouvelle ligne de données
df_new = pd.DataFrame([new_data])

# Ajoute un identifiant à la nouvelle ligne
df_new['id'] = last_id + 1

# Ajoute la nouvelle ligne au DataFrame existant
df_final = pd.concat([df_existing, df_new], ignore_index=True)

# Enregistre le DataFrame final dans le fichier CSV, sans index pour ne pas ajouter une colonne d'index non désirée
df_final.to_csv(filename, index=False)

print("Données ajoutées et enregistrées avec succès.")
```
